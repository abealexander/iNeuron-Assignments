In this assignment we have to find the frequency of words in a  webpage. We can use urllib and BeautifulSoup to extract text from  webpage.

```python
# importing libraries
from bs4 import BeautifulSoup 
import urllib.request 
import nltk
import string
```

```python
# opening the url
try:
    response = urllib.request.urlopen('http://php.net/') 
except: 
    print("Incorrect URL. Please check URL again.")

# reading the response    
html = response.read() 
```

```python
# converting to html
soup = BeautifulSoup(html,"html5lib")
```

```python
# only retaining important text
text = soup.find_all(text=True)
output = ''
blacklist = [
    '[document]',
    'noscript',
    'header',
    'html',
    'meta',
    'head',
    'input',
    'script',
]

for t in text:
    if t.parent.name not in blacklist:
        output += '{}'.format(t)

print(output)
```

```python
# punctuations saved
punct = string.punctuation
punct

# lets remove punctuations and perform tokenization
cleaned_tokens = []
for word in nltk.word_tokenize(output):
    if word not in punct:
        cleaned_tokens.append(word)
        
print(cleaned_tokens[0:10])
```

```python
# using counter to count cleaned_tokens
from collections import Counter
Counter(cleaned_tokens).most_common()
```
