In this assignment we have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference.
The raw image is available in spicy.misc package with the name face.

```python
import numpy as np
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt
```

```python
# loading the image
face = misc.face(gray=True)

# row and column variable
rows = face.shape[0]
cols = face.shape[1]

# checking the dimensions
face.shape
```

```python
# total pixels
face.size
```

```python
# datatype
face.dtype
```

```python
# visualizing the uncompressed image
plt.gray()
plt.imshow(face)
plt.show()
```

```python
# converting image to 1 d array
flat = face.reshape(rows*cols,1)
```

```python
# using kmeans to fit image into 5 clusters
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 10)
kmeans.fit(flat)
```

```python
# replacing each pixel value with corresponding cluster centroid
compressed_image = kmeans.cluster_centers_[kmeans.labels_]

# reshaping the image to original dimension
compressed_image = compressed_image.reshape(rows, cols)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].imshow(face)
axes[1].imshow(compressed_image)
axes[0].set_title("Raw Image", fontsize = 15)
axes[1].set_title("Compressed Image", fontsize = 15)
plt.gray()
fig.tight_layout()
```
