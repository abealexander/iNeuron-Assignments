### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import cufflinks as cf
import plotly.express as px
from sklearn.decomposition import PCA
```

```python
py.offline.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
pca = PCA(n_components=3)
new_data = pca.fit_transform(iris.iloc[:,0:4])
principal_x = pd.DataFrame(new_data,columns=['PC-1','PC-2','PC-3'])
```

```python
fig = px.scatter_3d(principal_x, x='PC-1', y='PC-2', z='PC-3', color=iris['species'])
fig.show()
```
